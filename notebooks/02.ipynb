```python
# ==========================================================
#  NOTEBOOK 02 — VERIFICAÇÃO DOS ARQUIVOS
# ==========================================================

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("BoardGames02").getOrCreate()

print("Carregando nodes.csv ...")
nodes = spark.read.csv("/app/dados/nodes.csv", header=True, inferSchema=True)
nodes.show(10, truncate=False)
nodes.printSchema()

print("Carregando edges.csv ...")
edges = spark.read.csv("/app/dados/edges.csv", header=True, inferSchema=True)
edges.show(10, truncate=False)
edges.printSchema()

print("Contagens:")
print(f"Nodes: {nodes.count()}")
print(f"Edges: {edges.count()}")

print("Notebook 02 finalizado com sucesso!")
```
