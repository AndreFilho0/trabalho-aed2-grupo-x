```python
import pandas as pd
import glob

# ===============================
# 1) Localizar os CSVs
# ===============================
pagerank_file = glob.glob("/app/dados/output_pagerank/part-*.csv")[0]
lpa_file = glob.glob("/app/dados/output_lpa/part-*.csv")[0]

# Ler os CSVs
pagerank_df = pd.read_csv(pagerank_file)
lpa_df = pd.read_csv(lpa_file)

# ===============================
# 2) Todos os jogos por PageRank (ordenado)
# ===============================
pagerank_sorted = pagerank_df.sort_values("pagerank_score", ascending=False)

print("=== Jogos por PageRank (todos) ===")
print(pagerank_sorted[["boardgame_id", "boardgame_title", "pagerank_score"]])

# Salvar tabela completa
pagerank_sorted.to_csv("/app/dados/pagerank_all.csv", index=False)

# ===============================
# 3) Comunidades LPA (todos os jogos)
# ===============================
# Agrupar por comunidade
lpa_grouped = lpa_df.groupby("community_label").agg({
    "boardgame_id": list,
    "boardgame_title": list
}).reset_index()

print("\n=== Comunidades LPA (todas) ===")
for _, row in lpa_grouped.iterrows():
    print(f"Comunidade {row['community_label']}: {len(row['boardgame_id'])} jogos")

# Salvar tabela completa
lpa_grouped.to_csv("/app/dados/lpa_grouped_all.csv", index=False)

print("\nTabelas completas salvas em /app/dados/pagerank_all.csv e /app/dados/lpa_grouped_all.csv")
```
